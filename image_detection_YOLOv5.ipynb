```python
import torch
import os

# Model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')
```

```python
# Images
dir = '/content/gdrive/My Drive/images/'
imgs = [dir + f for f in ('1.jpg', '2.jpg', '3.jpg','4.jpg')]  # batch of images

# Inference
results = model(imgs)
results.print()  # or .show(), .save()
exec_path = '/content/gdrive/My Drive/images/'
results.save(os.path.join(exec_path,'YOLOv5_detected'))
```

```python
from IPython.display import display, Image
dir2 = '/content/gdrive/My Drive/images/YOLOv5_detected/'
imgs_output = [dir2 + f for f in ('1.jpg', '2.jpg', '3.jpg','4.jpg')]
for items in imgs_output:
  # print(items)
  # print(type(items))
  display(Image(filename=items))



    
```
